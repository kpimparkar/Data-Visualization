## Abstract 
> Water scarcity is the lack of fresh water resources to meet water demand. 
It affects every continent and was listed in 2019 by the World Economic Forum as
one of the largest global risks in terms of potential impact over the next decade.
Two-thirds of the global population (4 billion people) live under conditions of 
severe water scarcity at least 1 month of the year. Half a billion people in 
the world face severe water scarcity all year round. Half of the worldâ€™s 
largest cities experience water scarcity.  
[Water scarcity Wiki](https://en.wikipedia.org/wiki/Water_scarcity)
 
Water is essential for all living beings. Human race relies on water for 
drinking needs, food, medicine, electricity, manufacturing among other needs. 
Access to clean water is a basic need for humans. Water sources and reservoirs 
have played vital role in the development of human cultures.  

Even though water plays such an important role in the existence and development 
of human beings, we are currently facing grave problems of water scarcity and 
drought risks. Communicating this  risk to wider audience is now of more 
importance than ever before. 

Science direct explains the problem of water scarcity by taking into 
consideration the impact of human water consumption, natural factors and 
current context of this problem:

> Water scarcity arises in situations where there is insufficient water to 
simultaneously support both human and ecosystem water needs (White, 2014). Most 
often this arises as a result of a basic lack of water (i.e., physical 
water scarcity), but it may also result from a lack of suitable infrastructure 
to provide access to what might otherwise be considered ample available water 
resources, which is referred to as economic water scarcity. Physical water 
scarcity may occur as a result of both natural phenomena (e.g., aridity, drought) 
as well as from human influences (e.g., desertification, water storage; Pereira 
et al., 2009; White, 2014), although these influences are often coupled. 
For example, the process of desertification often commences as a result of water
overuse during periods of temporary drought; droughts are more common in arid 
regions (McMahon et al., 1992).  
> [Water scarcity - Science Direct](https://www.sciencedirect.com/topics/earth-and-planetary-sciences/water-scarcity)
 
This visualization project aims to convey these risks to the audience to create 
awareness about water shortage and the importance of water conservation. The 
project will visualize the water scarcity data from various sources to 
demonstrate the adverse conditions of water shortage and drought. 
 
## Motivation
Around 71% of Earth's surface is covered by water and there are huge water 
reservoirs such as rivers, lakes, glaciers around us. But still we are facing 
water scarcity. Countries such as Australia, Brazil, India, China, Syria have 
faced very severe impact of drought.[[2]](https://www.wri.org/blog/2015/06/global-tour-7-recent-droughts)
   
> A mere 0.014% of all water on Earth is both fresh and easily accessible. Of the remaining water, 97% is saline and 
a little less than 3% is hard to access. Technically, there is a sufficient amount of freshwater on a global scale. 
However, due to unequal distribution (exacerbated by climate change) resulting in some very wet and some very dry 
geographic locations, plus a sharp rise in global freshwater demand in recent decades driven by industry, humanity is facing a water crisis. Demand is expected to outstrip supply by 40% in 2030, if current trends continue.   
[Water scarcity Wiki](https://en.wikipedia.org/wiki/Water_scarcity)

Water scarcity and drought have huge ecological and sociological impact. 
These conditions affect various aspects of human life, such as[[3]](http://www.eschooltoday.com/global-water-scarcity/effects-of-water-shortage.html):  

**Health**  
Water scarcity forces the community to use less clean or even contaminated 
water for drinking, cooking. This has direct impact on overall health of 
individuals of the society. Water is important for cleaning process as well. 
Shortage of water results in sanitation and cleanliness related problems.    

**Food production**  
Water is used in irrigation of crops. Additionally, water is used in processing 
and production of packaged food. Water scarcity has direct impact on these 
processes resulting in lower food production.  

**Poverty**  
In societies where agriculture is the main industry providing jobs and income
 to individuals, the impact of water shortage is very threatening. Community 
 spends most of the time in search of clean water than on betterment of the 
 society.    

One of the factor that motivated me to perform analysis on the topic of water
 scarcity and droughts is the socioeconomic impact this problem has on 
 agriculture based countries like India. Agriculture sector is the only 
 livelihood to the two-third of India's population which gives employment to 
 the  57% of work force and is a raw material source to large number of 
 industries. Impact of droughts on the lives of Indian farmers is so deep 
 that it has resulted in suicides of farmers at epidemic proportions.
 Suicides among farmer community is a major concern in various state of India. 
 Approximately 3.4 million cotton farmers occupy a small region in India 
 call "Vidarbha" and 95% of them struggle with massive debt [[4]](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2738339/). Incidence
  of farmers ending their lives in this region had hit epidemic like proportions recently.
> In a country of 70 million farmers, it is 10 in every 100,000 farmers
committing suicide. This is higher than the total national suicide rate. The 
number of farmers committing suicide in India is more than twice of the total
 number of suicides being committed in the top 100 countries on the suicide 
 list! This indeed is worrying factor. The Government's measures including 
 waiving off loans, construction of dams, and other assisting measures have 
 not produced positive results so far.  
> [Suicides among farmers in India](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2738339/)

This is visualization project is an effort to emphasize on the *need for 
action*. Every important decision made by us depends on relevant data. In 
order to convey the wide spread nature of the water scarcity and drought 
risks to global leaders, it is necessary that this data is represented objectively. In order to represent the data objectively, this project relies on the Aqueductâ„¢ water risk framework provided by World  Resources Institute [[1]](https://www.wri.org/aqueduct).

> WRIâ€™s Aqueductâ„¢ information platform compiles advances in hydrological modeling,   
> remotely sensed data, and published data sets into a freely accessible online platform. 

Aqueduct water risk framework transforms various data sources into water risk
indicators. This project focuses on utilizing these indicators to convey the
vast spread of water scarcity. WRI provides an online tool which explores 
this data and visualizes it in a geospatial manner.

## Related work
Various resources are available where the current state of water scarcity is 
visualized in very effective manners. These plots do commendable jobs in 
conveying the risks we are facing today due to water shortage. Couple of such
visualizations are mentioned as below:

### U.S. Drought Monitor
#### Overview
This is geospatial visualization of current drought condition in the U.S. 
It displays the information about drought types, intesity of the drought 
and source of the data. The type and intensity of the drought is overlayed 
on the map of U.S. indicating conditions at various geographical locations in
the U.S.  
#### Explanation of the data
This visualization uses drought type, drought intensity at various geographic 
locations in the U.S. It uses shades of red and yellow to convey the intensity
of drought. The data is collected by government agencies monitoring the drought 
conditions.  
#### Visualization technique and design
This geospatial visualization clearly conveys following data:
- which states are affected by drought situation
- what is the severity of drought in these states
- if this is a log term problem or a short term problem faced by the state.  
Colors used in this visualization are easily distinguishable. They portray 
affected areas clearly. Mention of all government agencies involved in data 
collection and the author gives credibility to the data represented in this 
plot. This visualization has included Alaska, Hawaii, Puerto Rico as well 
indicating that those islands are also affected by the drought conditions. 
Time stamp of the visualization helps to convey how recent the data is. And 
the visualization has also provided viewers a way to access additional 
information via a website.

### Global view of duration of water scarcity
#### Overview
This is a visualization of duration of water scarcity in the world. It uses
map of the world and plots indicators of scarcity with red and green colors.
Purpose of the visualization appears to indiate which parts of the world 
are affected by water scarcity and the intensity of this scarcity. Visualization 
includes a scale to relate the color used with the period for which the region 
faced water scarcity. It succeeds in providing an overview of the water scarcity
in various parts of the world.
#### Explanation of the data
The visualization uses map of the world. Data visualized on this map is count of 
months when water scarcity was greater than 100%. This count is then converted to 
categorical values by choosing ranges. Each range is assigned a color which is then 
added as a layer on top of the global map. The data from the year 2016 is used for this 
visualization.
#### Visualization technique and design
Main medium of communication in this visualization is color. There is a scale
which indicates range of drought affected months represented by correponding
color. It successfully shows us which continents are affected the most by 
water scarcity.  
This visualization gives overall idea of spread of water scarcity, but it 
fails to provide us numbers behind the visual. It would benefit a lot by 
including some interactivity. Due to small scale of the graph, we don't get 
any idea about the condition of drought in a particular country.
The visualization uses red and green colors which are not visible to color blind people.
It doesn't clearly explain what does 'water scarcity greater than 100%' mean. It also 
fails to convey if this is a  permanent situation at the given countries or a 
temporary situation. Display of time trend would give us better idea of the issue.

## Objectives
Main objective of this project is to demonstrate how wide spread the problem 
of water scarcity and drought is. Instead of relying on just colors and 
geospatial visualizations, the project aims at showcasing quantified 
attributes to convey the impact of water scarcity. This approach will help 
perform a comparative analysis of the problem faced by the world. Such 
analysis will also generate healthy competition between communities to 
perform better to conserve water and control the adverse effect of water 
scarcity. 

WRI classifies the water risk in three major categories [[5]](https://wriorg.s3.amazonaws.com/s3fs-public/aqueduct-30-updated-decision-relevant-global-water-risk-indicators_0.pdf)
* Physical Risk Quantity
* Physical Risk Quality
* Regulatory and Reputational Risk  

This project focuses on attributes of "Physical Risk Quantity". Project will 
include following visualizations:  

* Analyze per capita usage of water across the world:
    Check if there is any relation between the per capita usage of water with
    the water scarcity. 
    * Water withdrawals per capita
    * Time trend of water withdrawals per capita
* Compare baseline water stress and depletion:
    Visualizations to compare magnitude of risk faced by different countries
    * Comparison of baseline water stress in various countries
    * Comparison of baseline water depletion in various countries
* Visualizing global coastal flood risks
    * To check if countries have extreme cases of flood risks and water scarcity
* Visualizing global drought risks
    * To demonstrate the increasing regions facing drought conditions

## Data sets and visualization methods
1) [Aqueduct Global Maps 3.0 Data](https://www.wri.org/resources/data-sets/aqueduct-global-maps-30-data)  
    This data set is used by the Aqueduct framework for their visual tool. 
    This data set is recent as of August 2019. The provided data will need 
    fair amount of reformatting. This will be done using Python pandas.
    Data set contains more than 250 attributes. Only relevant attributes will
    be extracted for visualizations.  
    The dataset has annual, monthly and projected data. This project will focus on annual and 
    projected data. Annual data will be used to represent severity of the water scarcity faced
    by the world in present. The projected dataset will be used to highlight how the situation
    might become worse if we do not act now.
    Some of the key attributes to be used for the visualizations are:
    
* Weighting Scheme  

|Attribute|Meaning|  
|-------------|-----|  
|def|Default                |
|agr|Agriculture            |
|che|Chemicals              |
|con|Construction Materials |
|elp|Electric Power         |
|fnb|Food & Beverage        |
|min|Mining                 |
|ong|Oil & Gas              |
|smc|Semiconductor          |
|tex|Textile                |

* Identifiers  

|Attribute | Meaning|
 |-----------|-----------|
 |gid_0		|ISO A3 country code based on GADM.|
 |name_0	|	National or political entity name based on GADM.|
 |name_1	|	Sub-national or political entity name based on GADM.|
 |area_km2|	area of the geometry in km2 (union of sub-basin, province and groundwater aquifer).|
 |drr|Drought risk indicator|   
 
* Indicators  

|Attribute| Meaning|  
|---|--------------------------|
|gtd|Groundwater table decline |
|bws|Baseline water stress     |
|bwd|Baseline water depletion  |

* Types  

|Attribute| Meaning|  
|---|--------------------------|
|_score|each indicator is mapped to a [0-5] scale.|
|_label|A label explaining the category of the indicator including threshold. e.g. "Extremely High (more than 1 in 100)".|
|_cat	|integer for each category [-1,4], can be used for visuals.|  

[Details of available attributes](https://github.com/wri/aqueduct30_data_download/blob/master/metadata.md)  
     
2) [Aqueduct Global Maps 2.0 Data](https://www.wri.org/resources/data-sets/aqueduct-global-maps-20)  
    Aqueduct global maps show static data in its tool. This is older data made 
    available by Aqueduct framework. Purpose of using this data is to compare
     water scarcity attributes reported in 3.0 version with 2.0 version. Attributes present in this dataset are similar to the attributes available for Aqueduct Global Maps 3.0, which will make the comparison possible. 
     
3) [Water withdrawals per capita](https://ourworldindata.org/water-use-sanitation#water-stress-and-scarcity)     
    This data set will be used to create time trend of per capita water 
    withdrawal for various countries. Additionally it will also compare this 
    per capita value across countries for 2016 data. 
    This data already is in a clean format. It contains the country ID, year of observation and the value of the water withdrawal per capita.
